```python
import pandas as pd
import datetime
import numpy as np
import sklearn as sk
import re
```

```python
data = pd.read_csv('holdout_set.csv', encoding = "ISO-8859-1")
```

```python
data.head()
```

```python
data['Len_Desc'] = data['Description'].str.len()
```

```python
data['Tagged_Count'] = data['Description'].str.count('@')
```

```python
data.head()
```

```python
data= data.rename(columns = {"Followers at Posting" : "NBA_Follower_Count"})
```

```python
data['DayOfWeek']= pd.to_datetime(data['Created']).dt.day_name()
```

```python
data.head()
```

```python
data['Created']= pd.to_datetime(data['Created'])
```

```python
def playoffDate(date): 
    if date > pd.Timestamp('2019-4-13'): 
        return 1 
    else: 
        return 0
```

```python
data['playOffDate']= data.Created.map(lambda a: playoffDate(a.tz_localize(None)))
```

```python
data['month'] = data.Created.map(lambda a: a.month)
```

```python
data.head()
```

```python
data['hour'] = data.Created.map(lambda a: a.hour)
```

```python
data.head()
```

```python
def timeOfDay(hour): 
    if hour < 6: 
        return 'Night' 
    elif hour <12: 
        return 'Morning'
    elif hour <18: 
        return 'Afternoon'
    else:
        return'Evening'
```

```python
data['timeOfDay']= data.hour.map(lambda a: timeOfDay(a))
```

```python
data.head()
```

```python
def proportionOfTarget(caption): 
    #In this case target will be all non normal text
    targetCount = 0
    if not caption: return 0
    lst = caption.split(' ')
    count = len(lst)
    if count == 0: return 0
    for word in lst: 
        if len(word) ==0: continue 
        if word[0] in ['@', '#']: targetCount += 1
        elif '??' in word: targetCount+= 1
    return float(targetCount)/float(count)
    
    
```

```python
data['propOfSpecialTexts'] = data.Description.map(lambda a: proportionOfTarget(a))
```

```python
data.head()
```

```python
data['shoutouts']= data.Description.map(lambda x: re.findall('@\w*', x))
data.head()
```

```python
allVals = {}
for listItem in data['shoutouts']:
    for item in listItem:
        if item.lower() not in allVals: allVals[item.lower()] = 0
        allVals[item.lower()] = allVals[item.lower()] + 1
```

```python
def tagLevel(tags):
    high = ['@kyrieirving', '@kingjames', '@stephencurry30', '@zo', '@easymoneysniper', '@spidadmitchell', '@jharden13', '@jaytatum0', '@dwyanewade', '@antdavis23', '@giannis_an34', '@karltowns', '@cp3', '@damianlillard', '@traeyoung', '@vicoladipo', '@lukadoncic', '@joelembiid', '@ygtrece']
    mid = ['@kevindurant', '@nbaontnt', '@lakers', '@demar_derozan', '@cavs', '@celtics' , '@nbatv', '@bensimmons', '@warriors', '@nyknicks', '@nbaonespn', '@timberwolves', '@atlhawks', '@russwest44', '@nbahistory', '@suns', '@washwizards', '@laclippers', '@hornets', '@pacers', '@miamiheat', '@houstonrockets', '@sacramentokings']
    low = ['@dallasmavs', '@sixers', '@bucks', '@raptors', '@trailblazers', '@orlandomagic', '@utahjazz', '@nbaallstar', '@pelicansnba', '@spurs', '@detroitpistons', '@brooklynnets', '@okcthunder', '@nuggets', '@nbasummerleague']
    
    if set(tags).intersection(set(high)):
        return 'High'
    elif set(tags).intersection(set(mid)):
        return 'Mid'
    elif set(tags).intersection(set(low)):
        return 'Low'
    else:
        return 'None'
```

```python
data['TagLevel'] = data.shoutouts.map(lambda a: tagLevel(a))
data.head()
```

```python
teams = ['@nuggets ','@spurs ', '@okcthunder ', '@pelicansnba ','@dallasmavs ', '@trailblazers ', '@utahjazz ',
        '@bucks ', '@sixers ','@raptors ','@laclippers ', '@miamiheat ', '@pacers ', '@houstonrockets ',
        '@timberwolves ', '@cavs ', '@warriors ','@celtics ','@lakers ','@sacramentokings ', '@orlandomagic ', '@detroitpistons ',
         '@brooklynnets ','@atlhawks ', '@memgrizz ', '@suns ', '@hornets ', '@washwizards ', '@nyknicks ', '@chicagobulls '
        ]
```

```python
def teamTagged(lst): 
    for tag in lst: 
        if tag in teams: 
            return 1 
    return 0 
```

```python
data['teamTagged']= data.shoutouts.map(lambda a: teamTagged(a))
```

```python
data.head()
```

```python
def otherTagged(lst): 
    for tag in lst: 
        if tag not in teams: 
            return 1 
    return 0 

data['otherTagged']= data.shoutouts.map(lambda a: otherTagged(a))
```

```python
top3 = ['@kyrieirving ','@kingjames ', '@stephencurry30 ']
bottom3 = ['@nuggets ','@spurs ','@okcthunder ' ]

def top3Tagged(lst): 
    for tag in lst: 
        if tag in top3: 
            return 1 
    return 0 

data['top3Tagged']= data.shoutouts.map(lambda a: top3Tagged(a))

def bottom3Tagged(lst): 
    for tag in lst: 
        if tag in bottom3: 
            return 1 
    return 0 

data['bottom3Tagged']= data.shoutouts.map(lambda a: bottom3Tagged(a))
```

```python
data['hashtags']= data.Description.map(lambda x: re.findall('#\w*', x))
data.head()
```

```python
data['hasHashtag']=data.hashtags.map(lambda x: x != [])
```

```python
def HashLevel(hashs):
    high = ['#KiaTipOff18', '#NBAPreseason', '#OneLastDance', '#NBAonABC', '#KiaTipOff17', '#BESTofNBA', '#nbapreseason', '#PhantomCam', '#NBABreakdown']
    mid = ['#NBAChinaGames', '#NBAFinals', '#NBAPlayoffs', '#NBAAllStar', '#NBAXmas', '#ThisIsYourTime', '#ThisIsWhyWePlay']
    low = ['#NBAKicks', '#TripleDoubleAlert', '#NBADraft', '#NBABDAY', '#NBAAwards', '#NBARooks', '#NBAMediaDay', '#TissotBuzzerBeater', '#NBASummer']
    
    if set(hashs).intersection(set(high)):
        return 'High'
    elif set(hashs).intersection(set(mid)):
        return 'Mid'
    elif set(hashs).intersection(set(low)):
        return 'Low'
    else:
        return 'None'
```

```python
data['HashLevel'] = data.hashtags.map(lambda a: HashLevel(a))
data.head()
```

```python
data['monthDateYear']= data.Created.map(lambda x: x.strftime('%Y-%m-%d'))
```

```python
x =data.groupby(['monthDateYear']).size()
x= x.reset_index()
data = data.merge(x, left_on='monthDateYear', right_on='monthDateYear')
data = data.rename(columns={0: 'postsInThatDay'})
```
